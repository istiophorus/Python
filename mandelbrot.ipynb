```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
import cmath
```

```python
def create_mandelbrot_set(width, height, max_iter):
    space = np.zeros((width, height))
    
    min_re = -3.0
    max_re = 1.0
    min_im = -2.0
    max_im = 2.0
    re_range = max_re - min_re
    im_range = max_im - min_im

    re = np.linspace(min_re, max_re, width)
    im = np.linspace(min_im, max_im, height)
    real, imaginary = np.meshgrid(re, im)
    
    c = real + 1j * imaginary
    
    z = np.zeros_like(c, dtype = np.complex128)
    img = np.zeros((width, height))

    for i in range(max_iter):
        mask = np.abs(z) < 2
        z[mask] = z[mask] * z[mask] + c[mask]
        #img += mask
        
    space = np.abs(z) > 2

    return space       
```

```python
def draw_madelbrot_set(width, height, max_iter):
    mset = create_mandelbrot_set(width, height, max_iter)
    img = mset * 255.0    
    fig = plt.figure(figsize=(16, 16))
    plt.imshow(img, cmap="gray")
```

```python
draw_madelbrot_set(1000, 1000, 100)
```
