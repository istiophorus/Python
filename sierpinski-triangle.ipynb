```python
import matplotlib.pyplot as plt
%matplotlib inline
import random as rnd
```

```python
P1 = (0.0, 0.0)
P2 = (10.0, 0.0)
P3 = (0.0, 10.0)
```

```python
vertices = [P1, P2, P3]
```

```python
def get_middle(p1, p2, ratio = 0.5):
    return ((p1[0] + p2[0]) * ratio, (p1[1] + p2[1]) * ratio)
```

```python
def generate_triangle_points(vertices, count, ratio):
    xx = []
    yy = []
    current = (1.0, 1.0)
    max_vertices_index = len(vertices) - 1

    for i in range(0, count):
        next_target_index = rnd.randint(0, max_vertices_index)
        next_target_point = vertices[next_target_index]

        middle = get_middle(current, next_target_point, ratio)
        xx.append(middle[0])
        yy.append(middle[1])
        current = middle
        
    return (xx, yy)
```

```python
def configure_subplot(ax):
    ax.set_facecolor('black')
    ax.axis('off')
    ax.set_axis_off()
    ax.add_artist(ax.patch)
    ax.patch.set_zorder(-1)  
```

```python
def draw_figure(xx, yy):
    fig = plt.figure(figsize = (20, 20), facecolor="black")
    configure_subplot(ax = plt.axes())
    plt.scatter(xx, yy, s = 1, c = "#b5e61d")    
    plt.show()
```

```python
xx, yy = generate_triangle_points(vertices, 50000, 0.5)
```

```python
draw_figure(xx, yy)
```
