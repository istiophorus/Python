```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cs = pd.read_csv("d:/WorkFolder/pdata.4/countries.csv", low_memory=False)
```

```python
cs.head()
```

```python
plt.title("GDP per capita vs. life expectancy")
plt.xlabel("log of GDP per capita")
plt.ylabel("Life expectancy")
plt.scatter(np.log10(cs.gdpPerCapita), cs.lifeExpectancy)
```

```python
years = sorted(cs.year.unique())
```

```python
for year in years:
    plt.figure()
    plt.title(str(year))
    plt.xlabel("log of GDP per capita")
    plt.ylabel("Life expectancy")
    plt.xlim(2, 6)
    plt.ylim(20,90)
    dat = cs[cs.year == year]
    plt.scatter(np.log10(dat.gdpPerCapita), dat.lifeExpectancy, 3)
    plt.savefig("d:/WorkFolder/pdata.4/" + str(year) + ".png")
```
