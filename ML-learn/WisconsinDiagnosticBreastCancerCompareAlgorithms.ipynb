```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
df = pd.read_csv("q:/Data.3/python-machine-learning-book-master/code/datasets/wdbc/wdbc.csv")
```

```python
def convert_symbol_to_value(x):
    if x == "M":
        return 0
    else:
        return 1
```

```python
df['diagnosis_num'] = df["Diagnosis"].apply(lambda x: convert_symbol_to_value(x))
```

```python
df.head()
```

```python
y = df["diagnosis_num"]
```

```python
X = df[["f1","f2","f3","f4","f5","f6","f7","f8","f9","f10","f11","f12","f13","f14","f15","f16","f17","f18","f19","f20","f21","f22","f23","f24","f25","f26","f27","f28","f29","f30"]]
```

```python
models = [
    RandomForestClassifier(n_estimators=20, random_state=0),
    KNeighborsClassifier(n_neighbors=7, algorithm='ball_tree'),
    LogisticRegression(random_state=4669),
    SVC(kernel='linear')
]
```

```python
def learn_and_evaluate(models, X, y, test_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=4669)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    
    for model in models:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        print(model)
        print(metrics.accuracy_score(y_test, y_pred))
```

```python
learn_and_evaluate(models, X, y, 0.30)
```
