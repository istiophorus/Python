```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
%matplotlib inline
```

```python
def bifurcation_1():
    ar = np.linspace(1.0, 4.0, 1000)
    y = np.empty(ar.shape)
    y.fill(0.5)
    yyy = []
    for q in range(0, 1050):
        y = y * (1 - y) * ar
        if q > 50:
            yyy.append(y)  

    res = []
    
    for yy in yyy:
        z = list(zip(ar, yy))
        res.extend(z)

    xx, yy = list(zip(*res))
    
    return (xx, yy)
```

```python
def bifurcation_2():
    ar = np.linspace(1.0, 4.0, 1000)
    xx = []
    yy = []
    for a in ar:
        y = 0.5
        for q in range(0, 1050):
            prev_y = y
            y = y * (1 - y) * a
            if q > 50:
                if y != prev_y:
                    xx.append(a)
                    yy.append(y)                  

    return (xx, yy)
```

```python
xx1, yy1 = bifurcation_2()
```

```python
xx2, yy2 = bifurcation_1()
```

```python
fig = plt.figure(figsize = (20, 16))
plt.scatter(xx1, yy1, s = 1)
```

```python
fig = plt.figure(figsize = (20, 16))
plt.scatter(xx2, yy2, s = 1)
```
